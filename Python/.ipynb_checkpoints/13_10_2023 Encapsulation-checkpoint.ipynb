```python
#Encapsulation

class Bank_Account:
    def __init__(self,balance):
        self._balance=balance
    def deposit(self,amount):
        self._balance+=amount
    def withdraw(self,amount):
        if amount>self._balance:
            print("insufficient balance")
        else:
            self._balance=self._balance-amount
    def get_balance(self):
        return self._balance
    
my_amount=Bank_Account(500)
my_amount.deposit(int(input("enter the deposit amount : ")))
my_amount.withdraw(int(input("enter the deposit amount : ")))
print("current balance is :",my_amount.get_balance())
        
    
```

```python
class Itvedant:
    def __init__(self, class_status="no"):
        self.class_status = class_status

    def set_class_status(self, status):
        if status == "yes" or status == "no":
            self.class_status = status
        else:
            print("Invalid input. Please provide 'yes' or 'no'.")

    def get_class_status(self):
        return self.class_status

my_class = Itvedant()

status = input("Is the class there? (yes/no): ")
my_class.set_class_status(status)

print("Current class status is:", my_class.get_class_status())
```

# Create a digital recipe book where users can add, search, and organize recipes. You can include features like ingredient conversion andÂ mealÂ planning. ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡

```python
class Recipe:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients

    def display_recipe(self):
        print(f"Recipe: {self.name}")
        print("Ingredients:")
        for ingredient in self.ingredients:
            print(f"- {ingredient}")
        self.cook()

    def cook(self):
        print("Follow the general cooking instructions.")

class BreakfastRecipe(Recipe):
    def cook(self):
        print("Cook the breakfast according to specific instructions.")

class DinnerRecipe(Recipe):
    def cook(self):
        print("Cook the dinner according to specific instructions.")

class RecipeBook:
    def __init__(self):
        self.recipes = []

    def add_recipe(self, recipe):
        self.recipes.append(recipe)

    def search_recipe(self, keyword):
        matching_recipes = []
        for recipe in self.recipes:
            if keyword.lower() in recipe.name.lower():
                matching_recipes.append(recipe)
        return matching_recipes

    def display_all_recipes(self):
        for recipe in self.recipes:
            recipe.display_recipe()

# Create recipe instances
breakfast = BreakfastRecipe("Scrambled Eggs", ["Eggs", "Salt", "Pepper"])
dinner = DinnerRecipe("Spaghetti Bolognese", ["Ground beef", "Tomato sauce", "Spaghetti"])

# Create a recipe book
recipe_book = RecipeBook()
recipe_book.add_recipe(breakfast)
recipe_book.add_recipe(dinner)

# Search for recipes and display them
keyword = input("Enter a keyword to search for recipes: ")
matching_recipes = recipe_book.search_recipe(keyword)
if matching_recipes:
    print("Matching Recipes:")
    recipe_book.display_all_recipes()
else:
    print("No matching recipes found.")
```

```python
class Graph:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Graph(new_x, new_y)

    def __sub__(self, other):
        new_x = self.x - other.x
        new_y = self.y - other.y
        return Graph(new_x, new_y)

    def __mul__(self, scalar):
        new_x = self.x * scalar
        new_y = self.y * scalar
        return Graph(new_x, new_y)

    def __truediv__(self, scalar):
        if scalar == 0:
            raise ValueError("Division by zero is not allowed")
        new_x = self.x / scalar
        new_y = self.y / scalar
        return Graph(new_x, new_y)

    def magnitude(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5

a = Graph(1, 2)
b = Graph(2, 3)

add_result = a + b 
sub_result = a - b  
mul_result = a * 2 
div_result = a / 2 
magnitude_a = a.magnitude()

print("Addition result: x =", add_result.x, ", y =", add_result.y)
print("Subtraction result: x =", sub_result.x, ", y =", sub_result.y)
print("Multiplication result: x =", mul_result.x, ", y =", mul_result.y)
print("Division result: x =", div_result.x, ", y =", div_result.y)
print("Magnitude of 'a':", magnitude_a)
```

# Create a system for managing student records, courses, and grades. Use classes to represent students, courses, and a database-like structure to store and manipulateÂ thisÂ data.ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡ğŸ‘‡

```python
# Define a class for students
class Student:
    # Initialize the student with name, id, and an empty list of courses
    def _init_(self, name, id):
        self.name = name
        self.id = id
        self.courses = []
    
    # Add a course to the student's list of courses
    def add_course(self, course):
        self.courses.append(course)
    
    # Remove a course from the student's list of courses
    def remove_course(self, course):
        self.courses.remove(course)
    
    # Get the student's name
    def get_name(self):
        return self.name
    
    # Get the student's id
    def get_id(self):
        return self.id
    
    # Get the student's list of courses
    def get_courses(self):
        return self.courses

# Define a class for courses
class Course:
    # Initialize the course with name, code, and an empty dictionary of grades
    def _init_(self, name, code):
        self.name = name
        self.code = code
        self.grades = {}
    
    # Add a grade for a student in the course
    def add_grade(self, student, grade):
        self.grades[student] = grade
    
    # Remove a grade for a student in the course
    def remove_grade(self, student):
        del self.grades[student]
    
    # Get the course's name
    def get_name(self):
        return self.name
    
    # Get the course's code
    def get_code(self):
        return self.code
    
    # Get the course's dictionary of grades
    def get_grades(self):
        return self.grades

# Define a class for a database-like structure to store and manipulate data
class Database:
    # Initialize the database with an empty list of students and an empty list of courses
    def _init_(self):
        self.students = []
        self.courses = []
    
    # Add a student to the database
    def add_student(self, student):
        self.students.append(student)
    
    # Remove a student from the database
    def remove_student(self, student):
        self.students.remove(student)
    
    # Add a course to the database
    def add_course(self, course):
        self.courses.append(course)
    
    # Remove a course from the database
    def remove_course(self, course):
        self.courses.remove(course)
    
    # Enroll a student in a course and add the course to the student's list of courses
    def enroll(self, student, course):
        student.add_course(course)
    
    # Unenroll a student from a course and remove the course from the student's list of courses
    def unenroll(self, student, course):
        student.remove_course(course)
    
    # Assign a grade to a student in a course and add the grade to the course's dictionary of grades
    def assign_grade(self, student, course, grade):
        course.add_grade(student, grade)
    
    # Change a grade for a student in a course and update the grade in the course's dictionary of grades
    def change_grade(self, student, course, grade):
        course.add_grade(student, grade)
    
    # Get the list of students in the database
    def get_students(self):
        return self.students
    
    # Get the list of courses in the database
    def get_courses(self):
        return self.courses
    
# Create some students and courses for testing purposes
s1 = Student("Alice",1)
s2 = Student("Bob", 2)
s3 = Student("Charlie", 3)

c1 = Course("Math", "MATH101")
c2 = Course("English", "ENGL101")
c3 = Course("Science", "SCI101")

# Create a database object and add some data to it
db = Database()
db.add_student(s1)
db.add_student(s2)
db.add_student(s3)

db.add_course(c1)
db.add_course(c2)
db.add_course(c3)

db.enroll(s1, c1)
db.enroll(s1, c2)
db.enroll(s2, c2)
db.enroll(s2, c3)
db.enroll(s3, c1)
db.enroll(s3, c3)

db.assign_grade(s1, c1, 90)
db.assign_grade(s1, c2, 80)
db.assign_grade(s2, c2, 85)
db.assign_grade(s2, c3, 75)
db.assign_grade(s3, c1, 95)
db.assign_grade(s3, c3, 70)

# Print some information from the database for verification purposes

print("Students in the database:")
for s in db.get_students():
    print(f"{s.get_name()} (ID: {s.get_id()})")

print("Courses in the database:")
for c in db.get_courses():
    print(f"{c.get_name()} (Code: {c.get_code()})")

print("Grades in each course:")
for c in db.get_courses():
    print(f"{c.get_name()} (Code: {c.get_code()}):")
    for s, g in c.get_grades().items():
        print(f"{s.get_name()} (ID: {s.get_id()}): {g}")
```
