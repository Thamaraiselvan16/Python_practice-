```python
# Population Growth Simulator:
def simulate_population_growth(initial_population, growth_rate, years):
    population = initial_population
    
    for year in range(1, years + 1):
        population = population + (population * growth_rate / 100)
        print(f"Year {year}: Population = {population:.0f}")

def main():
    print("Population Growth Simulator")
    initial_population = int(input("Enter the initial population: "))
    growth_rate = float(input("Enter the growth rate (in percentage): "))
    years = int(input("Enter the number of years: "))
    
    simulate_population_growth(initial_population, growth_rate, years)

if __name__ == "__main__":
    main()
```

```python
# Ecosystem Simulation:
import random

class Rabbit:
    def __init__(self, population):
        self.population = population

    def reproduce(self):
        self.population *= 2

    def get_eaten(self, wolves):
        eaten = min(self.population, wolves.population)
        self.population -= eaten
        wolves.population -= eaten

class Wolf:
    def __init__(self, population):
        self.population = population

    def reproduce(self):
        self.population *= 1.5

def simulate_ecosystem(rabbits, wolves, time_steps):
    for step in range(1, time_steps + 1):
        print(f"Time Step {step}:")
        print(f"Rabbits: {rabbits.population}")
        print(f"Wolves: {wolves.population}")

        rabbits.reproduce()
        wolves.reproduce()
        rabbits.get_eaten(wolves)

        print()

def main():
    initial_rabbit_population = random.randint(50, 100)
    initial_wolf_population = random.randint(10, 20)
    time_steps = 10

    rabbits = Rabbit(initial_rabbit_population)
    wolves = Wolf(initial_wolf_population)

    simulate_ecosystem(rabbits, wolves, time_steps)

if __name__ == "__main__":
    main()
```

```python
# Hospital Emergency Room Simulator:
import random
import time

class Patient:
    def __init__(self, name, condition):
        self.name = name
        self.condition = condition

class EmergencyRoom:
    def __init__(self):
        self.patients = []

    def admit_patient(self, patient):
        self.patients.append(patient)
        print(f"{patient.name} admitted with {patient.condition}.")

    def treat_patients(self):
        for patient in self.patients:
            print(f"Treating {patient.name} with {patient.condition}.")
            time.sleep(random.uniform(0.5, 2))
            print(f"{patient.name} treated and discharged.")
        self.patients.clear()

def main():
    emergency_room = EmergencyRoom()

    print("Hospital Emergency Room Simulator")

    while True:
        print("\nOptions:")
        print("1. Admit patient")
        print("2. Treat patients")
        print("3. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            name = input("Enter patient's name: ")
            condition = input("Enter patient's condition: ")
            patient = Patient(name, condition)
            emergency_room.admit_patient(patient)
        elif choice == 2:
            if emergency_room.patients:
                print("\nTreating patients:")
                emergency_room.treat_patients()
            else:
                print("No patients to treat.")
        elif choice == 3:
            print("Exiting the Emergency Room Simulator.")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
# Agricultural Crop Growth Simulator:
class Crop:
    def __init__(self, name, growth_rate, max_height):
        self.name = name
        self.growth_rate = growth_rate
        self.max_height = max_height
        self.height = 0
        self.days_grown = 0

    def grow(self):
        self.height += self.growth_rate
        self.days_grown += 1

    def harvest(self):
        self.height = 0
        self.days_grown = 0
        print(f"{self.name} has been harvested.")

def simulate_crop_growth(crop, days_to_simulate):
    for day in range(1, days_to_simulate + 1):
        crop.grow()
        print(f"Day {day}: {crop.name} height = {crop.height:.2f} cm")

def main():
    print("Agricultural Crop Growth Simulator")

    crop_name = input("Enter the name of the crop: ")
    growth_rate = float(input("Enter the growth rate per day (in cm): "))
    max_height = float(input("Enter the maximum height of the crop (in cm): "))
    days_to_simulate = int(input("Enter the number of days to simulate: "))

    crop = Crop(crop_name, growth_rate, max_height)
    simulate_crop_growth(crop, days_to_simulate)
    crop.harvest()

if __name__ == "__main__":
    main()
```

```python
# set alaram
import time

def set_alarm(alarm_time):
    print(f"Alarm set for {alarm_time}.")
    while True:
        current_time = time.strftime("%H:%M:%S")
        if current_time == alarm_time:
            print("Time to wake up!")
            break
        time.sleep(1)

def main():
    print("Alarm Clock")
    print("1. Set Alarm")
    print("2. Exit")
    
    choice = input("Select an option (1/2): ")
    
    if choice == "1":
        alarm_time = input("Enter alarm time (HH:MM:SS): ")
        set_alarm(alarm_time)
        
    elif choice == "2":
        print("Goodbye!")
        
    else:
        print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
# Guess the Number Game
import random

def guess_the_number(secret_number):
    tries = 0
    while True:
        guess = int(input("Guess the number: "))
        tries += 1
        
        if guess < secret_number:
            print("Too low. Try again.")
        elif guess > secret_number:
            print("Too high. Try again.")
        else:
            print(f"Congratulations! You guessed the number {secret_number} in {tries} tries.")
            break

def main():
    lower_limit = 1
    upper_limit = 100
    secret_number = random.randint(lower_limit, upper_limit)
    
    print("Welcome to Guess the Number Game!")
    print(f"I'm thinking of a number between {lower_limit} and {upper_limit}.")
    
    guess_the_number(secret_number)

if __name__ == "__main__":
    main()
```

```python
# card game
import random

def create_board(size):
    numbers = list(range(1, size // 2 + 1)) * 2
    random.shuffle(numbers)
    board = [numbers[i:i+size] for i in range(0, size, size // 2)]
    return board

def display_board(board, revealed):
    for row in range(len(board)):
        for col in range(len(board[row])):
            if revealed[row][col]:
                print(board[row][col], end="\t")
            else:
                print("?", end="\t")
        print()

def main():
    size = int(input("Enter the size of the board (even number): "))
    if size % 2 != 0:
        print("Size must be an even number.")
        return

    board = create_board(size)
    revealed = [[False] * size for _ in range(size)]
    pairs_found = 0

    while pairs_found < size * size // 2:
        display_board(board, revealed)

        while True:
            try:
                row1 = int(input("Enter the row of the first card: "))
                col1 = int(input("Enter the column of the first card: "))
                row2 = int(input("Enter the row of the second card: "))
                col2 = int(input("Enter the column of the second card: "))
                if row1 < 0 or row1 >= size or col1 < 0 or col1 >= size or row2 < 0 or row2 >= size or col2 < 0 or col2 >= size:
                    raise ValueError("Invalid input. Row and column indices must be within the board's dimensions.")
                break
            except ValueError as e:
                print(e)

        if board[row1][col1] == board[row2][col2]:
            print("Match!")
            revealed[row1][col1] = True
            revealed[row2][col2] = True
            pairs_found += 1
        else:
            print("No match. Try again.")

    print("Congratulations! You've matched all pairs.")

if __name__ == "__main__":
    main()
```

```python
# snake game
import random
import curses

def main(stdscr):
    curses.curs_set(0)
    stdscr.nodelay(1)
    stdscr.timeout(100)

    sh, sw = stdscr.getmaxyx()
    w = stdscr.subwin(sh, sw, 0, 0)
    w.keypad(1)

    snk_x = sw // 4
    snk_y = sh // 2
    snake = [
        [snk_y, snk_x],
        [snk_y, snk_x - 1],
        [snk_y, snk_x - 2]
    ]

    food = [sh // 2, sw // 2]
    w.addch(food[0], food[1], curses.ACS_PI)

    key = curses.KEY_RIGHT

    while True:
        next_key = w.getch()
        key = key if next_key == -1 else next_key

        if (
            snake[0][0] in [0, sh] or
            snake[0][1] in [0, sw] or
            snake[0] in snake[1:]
        ):
            curses.endwin()
            quit()

        new_head = [snake[0][0], snake[0][1]]

        if key == curses.KEY_DOWN:
            new_head[0] += 1
        if key == curses.KEY_UP:
            new_head[0] -= 1
        if key == curses.KEY_LEFT:
            new_head[1] -= 1
        if key == curses.KEY_RIGHT:
            new_head[1] += 1

        snake.insert(0, new_head)

        if snake[0] == food:
            food = None
            while food is None:
                nf = [
                    random.randint(1, sh - 1),
                    random.randint(1, sw - 1)
                ]
                food = nf if nf not in snake else None
            w.addch(food[0], food[1], curses.ACS_PI)
        else:
            tail = snake.pop()
            w.addch(tail[0], tail[1], ' ')

        w.addch(snake[0][0], snake[0][1], curses.ACS_CKBOARD)

curses.wrapper(main)
```
